```python
import zarr
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np

OUTPUT_ROOT = Path("output3")
FINAL_ZARR_PATH = OUTPUT_ROOT / "final_result.zarr"

STEP = 4  # 원본 그대로 보기

# Zarr load
z_res = zarr.open(str(FINAL_ZARR_PATH), mode='r')

# 5D 배열 형태: (t, c, z, y, x)
arr = z_res["0"]

# 원하는 슬라이스 즉시 NumPy 로 변환 (Dask 없음)
img = np.asarray(arr[0, 0, 0])   # 2D 7400×5600

# 다운샘플링 적용
img_view = img[::STEP, ::STEP]

plt.figure(figsize=(15, 15))
plt.imshow(img_view, cmap="gray")
plt.title(f"Final Deconvolved Image\nShape: {img_view.shape}")
plt.axis("off")
plt.tight_layout()
plt.show()
```
